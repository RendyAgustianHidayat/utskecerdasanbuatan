```python
import json

# Daftar perusahaan
companies = [
    {"name": "Company 1"},
    {"name": "Company 3"},
    {"name": "Company 2"}
]

# Mengurutkan perusahaan berdasarkan nama dalam urutan terbalik
companies.sort(key=lambda x: x["name"], reverse=True)

# Menampilkan hasil dalam format JSON yang disesuaikan
display_text = "[{" + "}, {".join(json.dumps(company, indent=4)[1:-1] for company in companies) + "}]"
print(display_text)
```

```python
# Daftar perusahaan dengan domain dan lokasi
companies = [
    {"name": "Company 1", "domain": "Retail", "location": "United States"},
    {"name": "Company 2", "domain": "Technology", "location": "United Kingdom"},
    {"name": "Company 3", "domain": "Healthcare", "location": "United States"}
]

# Menampilkan nilai domain untuk setiap perusahaan dalam urutan nomor yang benar
for company in companies:
    print(f"{company['name']}: {company['domain']} ({company['location']})")
```

```python
# Data perusahaan dengan domain dan karyawan
companies = [
    {"domain": "Retail", "employees": ["John Doe", "Robert White", "Ying Han Tan"]},
    {"domain": "Technology", "employees": ["Tom Smith", "Rubby Lee"]},
    {"domain": "Healthcare", "employees": ["Kenneth Ng", "Andrew Sebastian"]},
]

# Membuat dictionary baru dengan domain sebagai kunci dan daftar karyawan sebagai nilai
employees_by_domain = {company["domain"]: company["employees"] for company in companies}

# Menampilkan hasil dalam format JSON
import json
print(json.dumps(employees_by_domain, indent=4))
```

```python
# Data perusahaan dengan domain, lokasi, dan karyawan
companies = [
    {"name": "Company 1", "domain": "Retail", "country": "United States", "employees": ["John Doe", "Robert White", "Ying Han Tan"]},
    {"name": "Company 2", "domain": "Technology", "country": "United Kingdom", "employees": ["Tom Smith", "Rubby Lee"]},
    {"name": "Company 3", "domain": "Healthcare", "country": "United States", "employees": ["Kenneth Ng", "Andrew Sebastian"]}
]

# Fungsi untuk mendapatkan daftar karyawan dengan perusahaan dan negara
def get_employees_with_country(companies):
    result = []
    for company in companies:
        for employee in company["employees"]:
            result.append({
                "full_name": employee,
                "company": company["name"],
                "country": company["country"]
            })
    return result

# Menjalankan fungsi dan menyimpan hasilnya
employees_with_country = get_employees_with_country(companies)

import json

# Menampilkan hasil dalam format JSON
print(json.dumps(employees_with_country, indent=4))
```

```python
# Data perusahaan dengan lokasi dan karyawan beserta kewarganegaraan mereka
companies = [
    {
        "name": "Company 1",
        "country": "United States",
        "employees": [
            {"full_name": "John Doe", "nationality": "Australia"},
            {"full_name": "Robert White", "nationality": "Australia"},
            {"full_name": "Ying Han Tan", "nationality": "Hong Kong"}
        ]
    },
    {
        "name": "Company 2",
        "country": "United Kingdom",
        "employees": [
            {"full_name": "Tom Smith", "nationality": "Hong Kong"},
            {"full_name": "Rubby Lee", "nationality": "United States"}
        ]
    },
    {
        "name": "Company 3",
        "country": "United Kingdom",
        "employees": [
            {"full_name": "Kenneth Ng", "nationality": "United States"},
            {"full_name": "Andrew Sebastian", "nationality": "United States"}
        ]
    }
]

# Fungsi untuk mendapatkan perusahaan dengan jumlah karyawan berdasarkan kewarganegaraan
def get_companies_with_employee_nationality(companies):
    result = []
    
    for company in companies:
        employee_nationality = {}
        
        for employee in company["employees"]:
            nationality = employee["nationality"]
            if nationality in employee_nationality:
                employee_nationality[nationality] += 1
            else:
                employee_nationality[nationality] = 1
        
        result.append({
            "company": company["name"],
            "employee_nationality": employee_nationality
        })
    
    return result

# Menjalankan fungsi dan menyimpan hasilnya
companies_with_nationality = get_companies_with_employee_nationality(companies)

# Menampilkan hasil dalam format JSON
import json
print(json.dumps(companies_with_nationality, indent=4))
```
